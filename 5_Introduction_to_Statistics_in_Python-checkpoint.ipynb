```python
# importing necessary libraries
import statistics
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# sample data
datasetx = [1, 13, 55, 72, 89, 55, 44, 4]
# calculating standard deviation using both statistics and numpy library
print(statistics.stdev(datasetx))
```

```python
print(np.std(datasetx))
```

```python
# As we can see that results are different from these libraries because degree of freedom needs to be specified for std. dev
# in numpy degree of freedom can be specified but it can't be done in statistics function
```

```python
# importing csv
players = pd.read_csv(r'players.csv')
```

```python
players
```

```python
players.tail() # shows last 5 entries
```

```python
# describing the data
print(players['salary'].describe())
print(players.describe())
```

```python
# calculating variance
print (players['salary'].var())
```

```python
# calculating correlation
column_1 = players['salary']
column_2 = players['rating']
correlation = column_1.corr(column_2)
print(correlation)
```

```python
# correlation matrix
print(players.corr()) # this is the callable correlation method with diagonals as 1.00
```

```python
# pandas offer different methods of correlation ['pearson', 'spearman', 'kendall'] or callable
# this is why it is important to read the documentation
# https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.corr.html
```

```python
# trying out various correlation types
print(players.corr(method = 'pearson'))
```

```python
print(players.corr(method = 'spearman'))
```

```python
print(players.corr(method = 'kendall'))
```

```python
# using non-numerical functions
print(players['teams'].value_counts())
```

```python
print(players['teams'].describe())
```

```python
# creating a boxplot
boxplot = players.boxplot(column = ['salary'])
print(players['salary'].describe())
```

```python
# boxplot with matplotlib
x = players['salary']
fig, ax = plt.subplots()
ax.boxplot(x, showmeans = True, meanline = True, medianprops = {'linewidth':5,'color':'purple'})
plt.show()
```

```python
# plotting correlation using scatterplot
print(players.corr(method = 'pearson'))
```

```python
ax = players.plot.scatter(x = 'salary', y = 'rating', color = 'DarkBlue', label = 'Player Salary vs Rating')
```

```python
# plaotting bar graph for ratings
ratingcount = players['rating'].value_counts()
print(ratingcount)
```

```python
ratingcount.plot.bar()
plt.xlabel('rating')
plt.ylabel('number of instances')
```

```python
# replace bar() with barh() to get horizontal plot. make sure to reverse the x and y axis labels
ratingcount.plot.barh()
plt.ylabel('rating')
plt.xlabel('number of instances')
```

```python
# creating a piechart for team representation
players.groupby('teams').size().plot(kind = 'pie')
plt.title('Team representation')
plt.ylabel('')
```

```python
# Applying all the methods to a new dataset 'listings'
listings = pd.read_csv (r'listings.csv')
listings.head()
```

```python
# describing price column
print(listings['price'].describe())
```

```python
# boxplot for price
listings.boxplot(column = ['price'])
```

```python
# looking for correlations in lsitings data
print(listings[['minimum_nights','number_of_reviews','reviews_per_month']].corrwith(listings['price']))
```

```python
print(listings[['price','number_of_reviews','reviews_per_month']].corrwith(listings['minimum_nights']))
```

```python
# breaking down by group
print(listings['room_type'].describe())
```

```python
# piechart with percentages
listings.groupby('room_type').size().plot(kind = 'pie', autopct = '%1.0f%%')
plt.title('Room Type')
plt.ylabel('')
```

```python
# Using scatterplot to see the room location distribution
# plot longitude and latitude
listings.plot.scatter(x = 'longitude', y = 'latitude', label = 'AirBnB Amsterdam', alpha = 0.1 )
```
